#Sample Map Reduce Program to Count the Words

```
# Load text file as RDD
text_file = sc.textFile("dbfs:/FileStore/tables/Input.txt")
print(text_file.collect())

# Map each line to a list of words
words = text_file.flatMap(lambda line: line.split(" "))
print(words.collect())
```

```
# Map each word to a tuple of (word, 1)
word_counts = words.map(lambda word: (word, 1))
print(word_counts.collect())
```

```
# Reduce by key to calculate total count for each word
word_counts = word_counts.reduceByKey(lambda x, y: x + y)
print(word_counts.collect())
```

```
# Sort by word count in descending order
sorted_word_counts = word_counts.sortBy(lambda x: x[1], ascending=True)
print(sorted_word_counts.collect())
```

```
# Output the results
for word, count in sorted_word_counts.collect():
    print("{}: {}".format(word, count))
```

#Sample Map Reduce Program to Count the Genome Sequence

```
#Sample Genome Dataset
# Load text file as RDD
text_file = sc.textFile("dbfs:/FileStore/tables/genome_updated-1.txt")

# Map each line to a list of words
words = text_file.flatMap(lambda line: line.split())
print(words.collect())
```

```
# Map each word to a tuple of (word, 1)
word_counts = words.map(lambda word: (word, 1))
print(words.collect())
```

```
# Reduce by key to calculate total count for each word
word_counts = word_counts.reduceByKey(lambda x, y: x + y)
print(words.collect())
```

```
# Sort by word count in descending order
sorted_word_counts = word_counts.sortBy(lambda x: x[1], ascending=True)
print(sorted_word_counts.collect())
```

```
# Output the results
l=[]
for word, count in sorted_word_counts.collect():
    print("{}: {}".format(word, count))
    l.append(count)
```

```
print(l)
```

```
from collections import Counter
import matplotlib.pyplot as plt

counting = l
tmp_count = Counter (counting)
word_count = []

for i in tmp_count:
    word_count.append ([i, tmp_count[i]])

plt.plot([row[0] for row in word_count], [row[1] for row in word_count])
plt.axis([0,4,0,20])
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```

```
import pandas as pd
word_count=pd.DataFrame(data=l,columns=['List'])
word_count.List.hist()
plt.xlabel("Words")
plt.ylabel("Frequency")
```

#Map Reduce Program to Count the Genome Sequence

```
# Large Genome Dataset
# Load text file as RDD
text_file = sc.textFile("dbfs:/FileStore/tables/genome_updated.txt")

#print(text_file)

# Map each line to a list of words
words = text_file.flatMap(lambda line: line.split())
#print(words.collect())

# Map each word to a tuple of (word, 1)
word_counts = words.map(lambda word: (word, 1))

# Reduce by key to calculate total count for each word
word_counts = word_counts.reduceByKey(lambda x, y: x + y)


# Output the results
l=[]
for word, count in word_counts.collect():
    print("{}: {}".format(word, count))
    l.append(count)
```
